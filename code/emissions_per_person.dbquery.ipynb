```sql
select 
county_state_name,
population,
cast(
  replace(`GHG emissions mtons CO2e`, ',', '')
  as double
  ) / nullif(
    cast(
    replace(population, ',', '')
    AS double
  ), 0
) as emissions_per_person
from dev_projects.databricks_data_vizualization.emissions_data
order by emissions_per_person desc
```
