```sql
with top10 as(
  select 
  state_abbr,
  sum(cast(replace(`GHG emissions mtons CO2e`, ',', '')as double))as total_emissions
  from dev_projects.databricks_data_vizualization.emissions_data
  group by state_abbr
  order by total_emissions desc
  limit 10
)
select 
sum(total_emissions) as top10_emissions,
(sum(total_emissions) / 
(select sum(cast(replace(`GHG emissions mtons CO2e`, ',', '')as double)) from dev_projects.databricks_data_vizualization.emissions_data)
) * 100 as top10_percentage
from top10
```
